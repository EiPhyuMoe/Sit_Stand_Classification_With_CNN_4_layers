```python
import tensorflow as tf
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model('/content/drive/MyDrive/Sit_Stand/cnn_model_with_custom_gen_conv_layer_4.h5')
```

```python
categories = ['Negative','Sit','Stand']
```

```python
# predicting images
img = tf.keras.utils.load_img('/content/drive/MyDrive/Sit_Stand/train/negative/DS_006_002_003_001_000076.jpg', target_size=(224, 128))
x = tf.keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = x/255
plt.imshow(x[0])
classes = model.predict(x)
classes = np.argmax(classes,axis=1)
print ('Class:', categories[int(classes)])
print (classes)
```
